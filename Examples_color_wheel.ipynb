# Changing saturation and luminosity settings

```python
step = 6
shades = 12
wheel1 = color_wheel(luminocity = True, saturation = False).draw_wheel(shades=shades,step=step)
wheel2 = color_wheel(luminocity = False, saturation = True).draw_wheel(shades=shades,step=step)
wheel3 = color_wheel(luminocity = True, saturation = True).draw_wheel(shades=shades,step=step)
wheel4 = color_wheel(grey_wheel=True).draw_wheel(step=step)
wheels = [wheel1,wheel2,wheel3,wheel4]
fig=plt.figure(figsize=(15,15))
i=0
for wheel in wheels:
    ax=fig.add_subplot(2,2,i+1)
    ax.imshow(wheel)
    ax.axis('off')
    i+=1
```

# Changing steps and shades

```python
wheel1 = color_wheel(luminocity = True, saturation = False).draw_wheel(shades=3,step=10)
wheel2 = color_wheel(luminocity = False, saturation = True).draw_wheel(shades=24,step=2)
wheel3 = color_wheel(luminocity = True, saturation = True).draw_wheel(shades=6,step=50)
wheel4 = color_wheel(grey_wheel=True).draw_wheel(step=100)
wheels = [wheel1,wheel2,wheel3,wheel4]
fig=plt.figure(figsize=(15,15))
i=0
for wheel in wheels:
    ax=fig.add_subplot(2,2,i+1)
    ax.imshow(wheel)
    ax.axis('off')
    i+=1
```
